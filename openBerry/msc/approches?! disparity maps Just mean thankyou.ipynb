```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
disps=np.load(open("disps_list.np",'r'))['arr_0']
wim2=disps[50]
#wim3=disps[56]

wim=disps[56]
#plt.imshow(wim)
plt.imshow(wim)
plt.show()
```

```python
plt.imshow(wim)
```

```python
wim2fl=np.flip(wim,0)
w=wim2fl#wim2#cv2.imread("temp.png")
imgheight=w.shape[0]
imgwidth=w.shape[1]
y1 = 0
M = imgheight//4
N = imgwidth//4
import time
t0=time.time()
i=0
frames=np.zeros((16,60,80))
xnyPairs_andMean=np.zeros((16,3))

for y in range(0,imgheight,M):
    for x in range(0, imgwidth, N):
        y1 = y + M
        x1 = x + N
        tiles = w[y:y+M,x:x+N]
        frames[i]=w[y:y+M,x:x+N]
        xnyPairs_andMean[i]=[x,y,np.mean(tiles)]
        i+=1
        #cv2.rectangle(w, (x, y), (x1, y1), (0, 255, 0))
        cv2.imwrite("save/" + str(x) + '_' + str(y)+".png",tiles)
print(time.time()-t0)
```

```python
#notice how the y axis start from the bottom here but from the top in images,this system works!
fig, ax = plt.subplots()
xx=xnyPairs_andMean
loc=np.argmax(xnyPairs_andMean[:,2])
x,y=xx[:,0],xx[:,1]
#np.flip(xx[:,1],0) # just so it matches image indexing
print("play Audio @ {0}".format((x[loc],y[loc])))
ax.scatter(x,y)
for i in range(16):
    if i==loc:
            ax.annotate(str(i)+"POINT", (x[i],y[i]))
    else:
            ax.annotate(i, (x[i],y[i]))
    
plt.plot()
```
